```python
import pandas
import numpy
```

```python
path = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data" 
```

```python
df = pandas.read_csv(path,header=None) #Reading the csv file that we will work on.
```

```python
headers = ["symboling","normalized-losses","brand","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns = headers
```

```python
df
```

```python
df.replace ("?",numpy.nan,inplace= True) # We changed ? missing values into NaN (not a number) values.
```

```python
df
```

```python
missing_data = df.isnull()
missing_data.head(5)  #True means there is a missing value,False means there is no missing value.
```

```python
avg_1 = df["normalized-losses"].astype("float").mean(axis=0) #Data Formatting,from object type to float
```

```python
df["normalized-losses"].replace(numpy.nan,avg_1,inplace= True) #Dealing with Missing Values
```

```python
df["normalized-losses"]
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace = True)

# reset index, because we droped two rows
df.reset_index(drop = True, inplace = True)
```

```python
df["price"]
```

```python
df.dtypes
```

```python
df[['normalized-losses']] = df[['normalized-losses']].astype('int')
df[['price']] = df[['price']].astype('int')                                #Data Formatting
df[['bore','stroke']]= df[['bore','stroke']].astype('float')
```

```python
df.dtypes
```

```python
df['price'] = df['price']/df['price'].max() #Data Normalization
```

```python
df['price']
```

```python
"""We see the column "fuel-type" has two unique values, "gas" or "diesel". 
Regression doesn't understand words, only numbers. To use this attribute in regression analysis, 
we convert "fuel-type" into indicator variables.We will use the pandas' method 
'get_dummies' to assign numerical values to different categories of fuel type."""
```

```python
dummy_value_1 = pandas.get_dummies(df['fuel-type'])
dummy_value_1.head()
```

```python
dummy_value_1.rename(columns={'diesel':'fuel-type-diesel','gas':'fuel-type-gas'},inplace=True)
dummy_value_1.head(10)
#We now have the value 0 to represent "gas" and 1 to represent "diesel" in the column "fuel-type". 
#We will now insert this column back into our original dataset.
```

```python
df.columns
```

```python
dummy_value_1.head(5)
```

```python
# Merging our data frame "df" and "dummy_variable_1" 
df = pandas.concat([df, dummy_value_1], axis=1)

# Dropping original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```

```python
dummy_value_2 = pandas.get_dummies(df["aspiration"])
dummy_value_2.head(5)
```

```python
dummy_value_2.rename(columns={'std':'aspiration-std','turbo':'aspiration-turbo'},inplace=True)
dummy_value_2.head(5)
```

```python
df= pandas.concat([df,dummy_value_2],axis=1)
df.drop("aspiration",axis=1,inplace=True)
```
